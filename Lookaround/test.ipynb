```python
import pandas as pd  
import seaborn as sns  
import matplotlib.pyplot as plt  
  
# #   
# rounds = [  
#     ['RandomHorizontalFlip', 'AutoAugment', 'RandomVerticalFlip', 'RandAugment', 'RandomRotation'],  
#     ['RandomHorizontalFlip', 'AutoAugment', 'RandomVerticalFlip', 'RandAugment', 'RandomRotation'],  
#     ['RandomHorizontalFlip', 'AutoAugment'],  
# ]  
rounds=mydata
#   
#   
unique_methods = sorted(list(set([method for round in rounds for method in round])))  
  
# 0  
heatmap_data = [[0] * len(unique_methods) for _ in range(len(rounds))]  
  
# 1  
for i, round_methods in enumerate(rounds):  
    for method in round_methods:  
        index = unique_methods.index(method)  
        heatmap_data[i][index] = 1  
  
# DataFrameseaborn  
df = pd.DataFrame(heatmap_data, columns=unique_methods).T
  
#   
plt.figure(figsize=(100, 10))  
sns.heatmap(df, annot=False, cmap='coolwarm', linewidths=0)  
  
#   
plt.title('Method Selections Over Rounds Heatmap')  
plt.ylabel('Rounds')  
plt.xlabel('Methods')  
  
#   
plt.show()
```
